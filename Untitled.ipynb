```python
import os
import matplotlib.pyplot as plt
import csv
import numpy as np
import matplotlib

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (6, 6)
```

```python
dataset_path = os.path.expanduser('~/Downloads/Demo')
```

```python
sample_paths = [os.path.join(dataset_path, path) for path in os.listdir(dataset_path)]

print(len(sample_paths))
```

```python
samples = []

all_labels = []

for path in sample_paths:
    with open(path) as f:
        r = csv.reader(f)
        points = []
        labels = []
        for y, x, z in r:
            points.append([float(x), float(y)])
            labels.append(z)
        
        points = np.array(points)
#         print(points[:, 0].std(), points[:, 1].std())
        labels = np.array(labels)
        samples.append([points, labels])
        all_labels.extend(labels)
 
labels_uniq = list(np.unique(all_labels))
print(labels_uniq)

for i in range(len(samples)):
    samples[i][1] = [labels_uniq.index(z) for z in samples[i][1]]

all_points = np.concatenate([s[0] for s in samples], 0)
all_labels = np.concatenate([s[1] for s in samples], 0)
plt.scatter(all_points[:, 0], all_points[:, 1], c=all_labels, s=1)    
plt.show()

for points, labels in samples[:10]:
    plt.scatter(points[:, 0], points[:, 1], c=labels)
    plt.show()
```
